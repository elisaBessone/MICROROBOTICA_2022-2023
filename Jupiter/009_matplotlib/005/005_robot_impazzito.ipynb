```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SECONDI = 2*60*60
```

```python
x_y = np.random.choice(["nord", "sud", "ovest", "est", ], SECONDI)
x_y
```

```python
x = [0]
y = [0]
posizione = (0,0)
for i in range(x_y.size):
    if x_y[i] == "est":
        x.append(x[-1] + 1)
        y.append(y[-1])
    elif x_y[i] == "ovest":
        x.append(x[-1] - 1)
        y.append(y[-1])
        
    elif x_y[i] == "nord":
        y.append(y[-1] + 1)
        x.append(x[-1])
        
    elif x_y[i] == "sud":
        y.append(y[-1] - 1)
        x.append(x[-1])
```

```python
fig, ax = plt.subplots(figsize = (20,10))
ax.scatter(x, y, marker = 'v', c = "green", s=5)
ax.plot(x,y, color = 'grey', linewidth=0.2)
plt.show()
```

#### Soluzione numpy

```python
movimenti = np.array([[1,0], [-1,0], [0,1], [0,-1]])
movimenti
```

```python
N_MOV = 7200
storia_movimenti = np.random.randint(0,4, N_MOV)
storia_movimenti
```

### INDICIZZAZIONE FANCY
indicizzazione per array di indici per creare un nuovo array


```python
array_coordinate = movimenti[storia_movimenti]    #tante righe quanto  lungo storia movimenti con associate le
                                                    #coordinate dei vari punti
#somma cumulativa --> sommo l'elemento precedente con quello successivo, modo sequenziale
posizioni = np.cumsum(array_coordinate, axis = 0)
posizioni
```

```python
x = posizioni[:,0]
y = posizioni[:,1]
```

```python
fig, ax = plt.subplots(figsize = (16,8))
ax.scatter(x, y, s = 5)
ax.plot(x, y, color = "green")

plt.show()
```

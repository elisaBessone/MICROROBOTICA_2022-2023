### Caricamento e visualizzazione immagini presenti nella cartella

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_immagini = 4
fig, ax = plt.subplots(4, 1, figsize = (20,10))
for i in range(n_immagini):
    immagine = plt.imread(f"img{i}.JPG")
    ax[i].imshow(immagine)
plt.show()
```

```python
n_immagini = 4
for i in range(n_immagini):
    fig, ax = plt.subplots(figsize = (16,8))
    immagine = plt.imread(f"img{i}.JPG")
    ax.imshow(immagine)
    plt.show()
    plt.close()
```

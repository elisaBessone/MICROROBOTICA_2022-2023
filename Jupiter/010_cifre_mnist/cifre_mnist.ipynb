```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import fetch_openml    #per scaricare dataset presenti in internet
# scarica i dati dalla rete (cella sotto)
```

```python
mnist = fetch_openml("mnist_784")
```

```python
mnist
```

```python
data_array = mnist['data'].values     #se dice che è un pandas metto .values
```

```python
type(data_array)
```

```python
data_array.shape         #ogni riga è un'immagine, 70000 immagini in scala di grigi
```

```python
#l'immagine che ho salvato è lunga 748, io devo ottenere una matrice 28*28 quindi uso replace
esempio_cifra = data_array[6542].reshape(28,28)      
esempio_cifra.shape
```

```python
fig, ax = plt.subplots(figsize = (16,8))
ax.imshow(esempio_cifra, cmap="Greys")
plt.show()
```

```python
# grafico che contiene 100 figure
n_rows = 5
n_cols = 5
fig, axs = plt.subplots(nrows=n_rows, ncols=n_rows, figsize = (8,8))

for i in range(n_rows):
    for j in range(n_cols):
        esempio_cifra = data_array[np.random.randint(0,70000)].reshape(28,28)      
        axs[i,j].imshow(esempio_cifra, cmap = "Greys")
        axs[i,j].axis('off')
plt.show()
```

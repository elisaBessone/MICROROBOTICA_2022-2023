#### Esercizio temperature globali variazione
https://ourworldindata.org/grapher/temperature-anomaly

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = open('temperature-anomaly.csv')
file = f.readlines()
f.close()
intestazione = file[0]
intestazione = intestazione.split(',')
intestazione[-1] = intestazione[-1].strip()
file = file[1:-1]
file
```

```python
anni = []
temp = []
anni_temp = []
for riga in file:
    if "Global" in riga:
        temp_anno = riga.split(',')
        temp_anno[-1] = temp_anno[-1].strip()

        anni.append(int(temp_anno[2]))
        temp.append(float(temp_anno[3]))
```

```python
anni = np.array(anni)
temp = np.array(temp)
```

```python

fig, axs = plt.subplots(nrows=3, ncols=1, figsize = (20,20))

axs[0].plot(anni, temp, label = "Variazione temperatura media", color = 'orange')

axs[0].set_title(f"VARIAZIONE ANOMALIA TEMPERATURA ANNI: {anni[0]} - {anni[-1]}", color = 'orange', fontsize = 20)
axs[0].set_xlabel("ANNI")
axs[0].set_ylabel("ANOMALIA")
axs[0].legend()

axs[1].scatter(anni, temp, marker = 'o', s = 5, label = "Variazione temperatura media", c = 'g')
axs[1].set_title(f"VARIAZIONE ANOMALIA TEMPERATURA ANNI: {anni[0]} - {anni[-1]}", color = 'g', fontsize = 20)
axs[1].set_xlabel("ANNI")
axs[1].set_ylabel("ANOMALIA")
axs[1].legend()

axs[2].hist(temp , label = "Variazione temperatura media", color = 'r', bins = 20)
axs[2].set_title(f"FREQUENZA ASSOLUTA ANOMALIA TEMPERATURA ANNI: {anni[0]} - {anni[-1]}", color = 'r', fontsize = 20)
axs[2].set_xlabel("anni")
axs[2].set_ylabel("frequenza assoluta")
axs[2].legend()


plt.show()
```

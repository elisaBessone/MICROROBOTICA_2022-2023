```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lato_matrice = 100
```

```python
matrice = np.random.randint(0,255, (lato_matrice,lato_matrice))
```

```python
matrice
```

```python
fig, axs = plt.subplots(nrows=2, ncols=1, figsize = (40,40))
axs[0].imshow(matrice, cmap = "hsv")
axs[1].hist(matrice.flatten(), bins = 255, color = 'r', ec = 'b')   #matrice.flatten() toglie una dimensione
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open("censimento2011_comuni_02-12-2022.csv")
file = f.readlines()
file = file[1:]
file
```

```python
datiComuni = {"COMUNE": [], "POPOLAZIONE_UOMINI" : [], "POPOLAZIONE_DONNE" : [], "POPOLAZIONE_TOTALE": []}
datiComuni
```

```python
for riga in file:
    comune = riga.split(";")
    comune[-1] = comune[-1].strip()
    
    if comune[1] not in datiComuni:
        datiComuni['COMUNE'].append(comune[1])
        datiComuni['POPOLAZIONE_UOMINI'].append(int(comune[2]))
        datiComuni['POPOLAZIONE_DONNE'].append(int(comune[3]))
        datiComuni['POPOLAZIONE_TOTALE'].append(int(comune[4]))
        
for campi in datiComuni:
    datiComuni[campi] = np.array(datiComuni[campi])
        
datiComuni
```

```python
fig, ax = plt.subplots(figsize = (16,30))

plt.show()
```

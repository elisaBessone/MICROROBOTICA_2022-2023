```python
import numpy as np
import matplotlib.pyplot as plt
```

### Esercizio 2, prezzi giornalieri di un'azione

```python
n_days = 200
np.random.seed(121)
prezzo_giornaliero = 100 + np.cumsum(np.random.normal(0.1, 1.0, n_days))
```

```python
prezzo_giornaliero = np.array(prezzo_giornaliero)
```

#### punto 1: massimi e minimi locali del prezzo

```python
indice_minimi = np.logical_and(prezzo_giornaliero < np.roll(prezzo_giornaliero,1), prezzo_giornaliero < np.roll(prezzo_giornaliero, -1))
indice_minimi
minimi_locali = prezzo_giornaliero[indice_minimi]
minimi_locali
```

```python
indice_massimi = np.logical_and(prezzo_giornaliero > np.roll(prezzo_giornaliero,1), prezzo_giornaliero > np.roll(prezzo_giornaliero, -1))
indice_massimi
massimi_locali = prezzo_giornaliero[indice_massimi]
massimi_locali
```

#### punto 2: per quanti giorni l'azione ha un prezzo maggiore di 105

```python
maggiore_105 = prezzo_giornaliero > 105
prezzo_maggiore_105 = prezzo_giornaliero[maggiore_105]
len(prezzo_maggiore_105)
```

#### punto 3: tracciare il grafico dei prezzi

```python
fig, ax = plt.subplots(figsize = (16,8)) 
ax.plot(prezzo_giornaliero)
ax.set_ylabel("Prezzo ($)")
ax.set_xlabel("Giorni")
ax.set_title("Serie storica del prezzo")
plt.show()
```

#### punto 4: massima variazione di prezzo (differenza tra giorno precedente e successivo)

```python
variazione = (np.roll(prezzo_giornaliero, -1) - prezzo_giornaliero)[:-1]   
variazione
```

```python
np.max(variazione)
```

```python
np.argmax(variazione)
```

#### punto 5: grafico a dispersione variabile prezzo rispetto al prezzo giornaliero

```python
fig, ax = plt.subplots(figsize = (16,8))
ax.scatter(variazione, prezzo_giornaliero[:-1], marker = 'o', s = 3)
ax.set_ylabel("Prezzo ($)")
ax.set_xlabel("Variazione ($)")
ax.set_title("Variazione prezzo rispetto al prezzo giornaliero")
plt.show()
```

### Esercizio 3, funzioni

```python
x = np.linspace(-3, 3, 1000)
```

```python
F1 = (np.exp(-x**2))
F2 = (np.exp(-x**2)) * np.sin(20*x)
F3 = (np.exp(-x**2)) * np.cos(20*x)
F4 = (np.sin(20*x)) / (20*x)
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (20,20))

axs[0,0].plot(x, F1)
axs[0,0].set_title("e^(-x^2)")
axs[0,1].plot(x, F2)
axs[0,1].set_title("e^(-x^2)*sin(20*x)")
axs[1,0].plot(x, F3)
axs[1,0].set_title("e^(-x^2)*cos(20*x)")
axs[1,1].plot(x, F4)
axs[1,1].set_title("sin(20*x)/(20*x)")


plt.show()
```

```python
fig, ax = plt.subplots(figsize = (30,15))

ax.plot(x, F1, label = "e^(-x^2)")
ax.plot(x, F2, label = "e^(-x^2)*sin(20*x)", linestyle = "-.")
ax.plot(x, F3, label = "e^(-x^2)*cos(20*x)", linestyle = "--")
ax.plot(x, F4, label = "sin(20*x)/(20*x)", linestyle = ":")
ax.set_title("Rappresentazione 3 funzioni", fontsize = 20)
ax.set_xlabel("x")
ax.set_ylabel("y")


plt.grid()
plt.legend()
plt.show()
```

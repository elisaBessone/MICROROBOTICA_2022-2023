```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open("emotion.csv")
file = f.readlines()
file = file[1:]
```

```python
emozioni = {"0":"angry", "1":"disgust", "2":"fear", "3":"happy", "4":"sad", "5":"surprise", "6":"neutral"}
```

```python
array_pixels = []
emotion = []
for riga in file:
    array_pixel = []
    riga = riga.split(",")
    riga[-1] = riga[-1].strip()
    pixel = riga[1].split(" ")
    for p in pixel:
        array_pixel.append(float(p))
    emotion.append(riga[0])
    array_pixels.append(np.array(array_pixel))
```

```python
array_pixel = np.array(array_pixel)
```

```python
array_pixels = np.array(array_pixels)
```

```python
n = 3
fig, axs = plt.subplots(nrows=3, ncols=3, figsize = (20,20))
num = 0
for i in range(n):
    for j in range(n):
        face = array_pixels[num].reshape(48,48)      
        axs[i,j].imshow(face)
        axs[i,j].axis('off')
        axs[i,j].set_title(emozioni[emotion[num]])
        num = num+1
        
plt.show()
```

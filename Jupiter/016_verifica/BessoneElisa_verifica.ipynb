### Esercizio 2, prezzi giornalieri di un'azione

```python
n_days = 200
np.random.seed(121)
prezzo_giornaliero = 100 + np.cumsum(np.random.normal(0.1, 1.0, n_days))
```

```python
prezzo_giornaliero = np.array(prezzo_giornaliero)
```

#### punto 1: massimi e minimi locali del prezzo

```python
indice_minimi = np.logical_and(prezzo_giornaliero < np.roll(prezzo_giornaliero,1), prezzo_giornaliero < np.roll(prezzo_giornaliero, -1))
indice_minimi
minimi_locali = prezzo_giornaliero[indice_minimi]
```

```python
indice_massimi = np.logical_and(prezzo_giornaliero > np.roll(prezzo_giornaliero,1), prezzo_giornaliero > np.roll(prezzo_giornaliero, -1))
indice_massimi
massimi_locali = prezzo_giornaliero[indice_massimi]
massimi_locali
```

#### punto 2: per quanti giorni l'azione ha un prezzo maggiore di 105

```python
maggiore_105 = prezzo_giornaliero > 105
prezzo_maggiore_105 = prezzo_giornaliero[maggiore_105]
prezzo_maggiore_105
```

#### punto 3: tracciare il grafico dei prezzi

```python
fig, ax = plt.subplots(figsize = (16,8)) 
ax.plot(prezzo_giornaliero)
plt.show()
```

#### punto 4: massima variazione di prezzo (differenza tra giorno precedente, corrente e successivo)

```python
variazione =  prezzo_giornaliero - np.roll(prezzo_giornaliero, -1)
variazione = variazione - np.roll(prezzo_giornaliero, -2)
variazione = -variazione
variazione
```

```python
fig, ax = plt.subplots(figsize = (16,8))
ax.plot(variazione[1:-1])
plt.show()
```

#### punto 5: grafico a dispersione variabile prezzo rispetto al prezzo giornaliero

variazione_giornaliero = prezzo_giornaliero - np.roll(prezzo_giornaliero, -1)  
variazione_giornaliero

```python
fig, ax = plt.subplots(figsize = (16,8))
ax.scatter(variazione, prezzo_giornaliero, marker = 'o', s = 3)
plt.show()
```

### Esercizio 3, funzioni

```python
x = np.linspace(-3, 3, 200)
print(x)
k = 2
```

```python
F1 = (np.exp(-x**2))
F2 = (np.exp(-x**2)) * np.sin(20*x)
F3 = (np.exp(-x**2)) * np.cos(20*x)
F4 = np.sin(20*x) / 20*x
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (20,20))

axs[0,0].plot(x, F1)
axs[0,0].set_title("e^(-x^2)")
axs[0,1].plot(x, F2)
axs[0,1].set_title("e^(-x^2)*sin(20x)")
axs[1,0].plot(x, F3)
axs[1,0].set_title("e^(-x^2)*cos(20x)")
axs[1,1].plot(x, F4)
axs[1,1].set_title("sin(20x)/20x")



plt.show()
```

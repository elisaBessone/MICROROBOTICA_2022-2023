```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open("emotion.csv")
file = f.readlines()
file = file[1:]
f.close()
```

```python
emozioni = {"0":"Angry", "1":"Disgust", "2":"Fear", "3":"Happy", "4":"Sad", "5":"Surprise", "6":"Neutral"}
```

```python
dati = {"emotion":[], "pixel":[]}
array_pixels = []
emotion = []
for riga in file:
    array_pixel = []
    
    #separo le due colonne
    colonne = riga.split(",")
    colonne[-1] = colonne[-1].strip()
    
    dati["emotion"].append(colonne[0])
    
    #separo la riga per estrarre i valori dell'immagine
    pixel = colonne[1].split(" ")
    #appendo nella lista i singoli valore per poi rappresentare la matrice
    for p in pixel:
        array_pixel.append(int(p))
    
    emotion.append(riga[0])
    array_pixels.append(np.array(array_pixel))
    
    dati["pixel"].append(np.array(array_pixel))
```

```python
array_pixel = np.array(array_pixel)
```

```python
array_pixels = np.array(array_pixels)
```

```python
n = 3
fig, axs = plt.subplots(nrows=n, ncols=n, figsize = (20,20))
num = 0
for i in range(n):
    for j in range(n):
        #face = array_pixels[num].reshape(48,48)
        face = dati["pixel"][num].reshape(48,48)
        axs[i,j].imshow(face, cmap="gray")
        axs[i,j].axis('off')
        axs[i,j].set_title(emozioni[dati["emotion"][num]], fontsize = 15)
        #axs[i,j].set_title(emozioni[emotion[num]]) 
        num = num+1

plt.suptitle("Rappresentazione immagini persone", fontsize = 20)
plt.show()
```

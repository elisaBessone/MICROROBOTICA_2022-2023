```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100000 #punti in un quadrato da -1 a 1
x = np.random.uniform(-1.,1.,n)
y = np.random.uniform(-1.,1.,n)
s = 0
```

```python
ipotenusa = np.sqrt(x**2+y**2)
nc = np.count_nonzero(s<=1)
colori = np.zeros(n)
colori[ipotenusa<=1] = 1
```

```python
fig, ax = plt.subplots(figsize = (8,8))
ax.scatter(x,y, marker="*", c=colori, cmap="gnuplot")
plt.show()
```
